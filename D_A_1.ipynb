Для каждого пассажира умножьте значение столбца Fare на 1.3, если класс его билета - 1 или 2, и на 1.1, если класс его
билета - 3. Сохраните результаты как новый столбец Fare_new. (При реализации расчета используйте любые альтернативы методу
iterrows)

```python
#SLOW
%%timeit
result = []

for _, row in df_pd.iterrows():
  if row.Pclass == 1 | row.Pclass == 2:
    res = row.Fare * 1.3
    result.append(res)
  elif row.Pclass == 3:
    res = row.Fare * 1.1
    result.append(res)
```

```python
#FASTER
%%timeit
df_pd['Fare_new'] = [
    row.Fare * 1.3 if row.Pclass == 1 or row.Pclass == 2 else
    row.Fare * 1.1 if row.Pclass == 3 else
    row.Fare
    for row in df_pd.itertuples()
    ]
```

np.float64: 64-bit floating point Range: approximately -1.8 × 10¹⁵ to 1.8 × 10¹⁵ with 15 decimal digits of precision

**First INT convertion**

```python
df['price'] = df['price'].astype(np.int32)        #max price 1.330000e+07 | max int32: 2.10000e+09 (ENOUGH)
df['area'] = df['area'].astype(np.int16)          #max area 16200 | max int16 32767 (ENOUGH)
df['bedrooms'] = df['bedrooms'].astype(np.int8)   #max bedrooms 6 | max int8 127 (ENOUGH)
df['bathrooms'] = df['bathrooms'].astype(np.int8) #max bathrooms 4 | max int8 127 (ENOUGH)
df['stories'] = df['stories'].astype(np.int8)     #max stories 4 | max int8 127 (ENOUGH)
df['parking'] = df['parking'].astype(np.int8)  #max parking 3 | max int8 127 (ENOUGH)
```

```python
df.info(memory_usage='deep')
```

**Transform YES and NO columns**

```python
#function easy understandable
def transf_yes_no(df):
    for col in df.columns:
        if df[col].dtype == 'object' and df[col].isin(['yes', 'no']).all():
            df[col] = df[col].map({'yes': 1, 'no': 0}).astype('int8')
    return df
```

```python
#1 line directly but not very understandable
df[df.columns[df.dtypes == 'object']] = df[df.columns[df.dtypes == 'object']].apply(
    lambda col: col.map({'yes': 1, 'no': 0}).astype('int8') if col.isin(['yes', 'no']).all() else col
)
```
